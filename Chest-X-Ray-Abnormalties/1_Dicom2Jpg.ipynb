```python
train_path="/content/drive/MyDrive/dataset/train_data/" 
output_path = '/content/drive/MyDrive/dataset/converted256/' 
train_image_file_list = os.listdir(train_path)
train_image_file_list
```

```python
IMAGE_SIZE=256
i = 0
for image_id in train_image_file_list :
    #fixed values
    i += 1
    voi_lut=True
    fix_monochrome = True 
    dicom=pydicom.dcmread(train_path+image_id)
    if voi_lut:
        data=apply_voi_lut(dicom.pixel_array,dicom)
    else:
        data=dicom.pixel_array
        
    if fix_monochrome and dicom.PhotometricInterpretation =="MONOCHROME1":
        data=np.amax(data)-data
        
    data=(((data-np.min(data))/np.max(data))*255).astype('u8').astype('f8')
    
    new_array = cv2.resize(data,(IMAGE_SIZE,IMAGE_SIZE)) #IMAGE_SIZE=512
    new_array = np.reshape(new_array, (IMAGE_SIZE, IMAGE_SIZE,1))
    new_array = np.asarray(new_array)
    img = array_to_img(new_array)
    img.save(output_path + image_id.replace('.dicom', '.jpg'), dtype='u16', quality=100)
    print('images_saved_number',i)
print('done')
```
